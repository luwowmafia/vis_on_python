

Постройте графики 2 функций:
- f(x) = sin(x)
- g(x) = 1 - cos(x)

Для x на отрезке [-2$\pi$; 2$\pi$]

Отметьте отдельными маркерами точки пересечения графиков. Оба графика должны проходить через точку пересечения при любом приближении (т.е. найдите точки пересечения с абсолютной точностью и добавьте как значения аргумента, так и значение функции в список по которому строится кривая).

*Примечание 1. Для поиска точек пересечения имеет смысл аналитически решить тригонометрическое уравнение*

    sin(x) = 1 - cos(x)
    
*Примечание 2. Чтобы решить это уравнение вам могут потребоваться формулы:
$$sin^{2}\alpha =\frac{1-cos2\alpha }{2}$$

$$sin 2\alpha  = 2 \cdot  sin \alpha \cdot  cos \alpha $$

* Подпишите оси и выведите заголовок

* Выведите осевые линии

```python
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
import math
import numpy as np


x = np.arange(-6.28, 6.28, 0.1)
def f(x):
    return np.sin(x)

def g(x):
    return 1-np.cos(x)

fig_f = go.Figure()
fig_f.update_yaxes(range=[-1.5, 1.5], zeroline=True, zerolinewidth=2, zerolinecolor='LightPink')
fig_f.update_xaxes(range=[-6.5, 6.5], zeroline=True, zerolinewidth=2, zerolinecolor='#008000')
fig_f.add_trace(go.Scatter(x=x, y=f(x), mode='lines+markers',  name='f(x)=sin(x)'))
fig_f.update_layout(legend_orientation="h",
                  legend=dict(x=.5, xanchor="center"),
                  hovermode="x",
                  title="График функции f(x)=sin(x)",
                  xaxis_title="x",
                  yaxis_title="y",  
                  margin=dict(l=0, r=0, t=40, b=0))
fig_f.update_traces(hoverinfo="all", hovertemplate="Аргумент: %{x}<br>Функция: %{y}")
fig_f.show()
```

```python
fig_g = go.Figure()
fig_g.update_yaxes(range=[-1.5, 3], zeroline=True, zerolinewidth=2, zerolinecolor='LightPink')
fig_g.update_xaxes(range=[-6.5, 6.5], zeroline=True, zerolinewidth=2, zerolinecolor='#008000')
fig_g.add_trace(go.Scatter(x=x, y=g(x), mode='lines+markers',name='g(x)=1-cos(x)',
                         marker=dict(color='LightSkyBlue', size=3, line=dict(color='MediumPurple', width=3))))
fig_g.update_layout(legend_orientation="h",
                  legend=dict(x=.5, xanchor="center"),
                  hovermode="x",
                  title="График функции g(x)=1-cos(x)",
                  xaxis_title="x",
                  yaxis_title="y",
                  margin=dict(l=0, r=0, t=40, b=0))
fig_g.update_traces(hoverinfo="all", hovertemplate="Аргумент: %{x}<br>Функция: %{y}")
fig_g.show()
```

```python
fig = go.Figure()
fig.update_yaxes(range=[-1.5, 3], zeroline=True, zerolinewidth=2, zerolinecolor='LightPink')
fig.update_xaxes(range=[-6.5, 6.5], zeroline=True, zerolinewidth=2, zerolinecolor='#008000')
fig.add_trace(go.Scatter(x=x, y=f(x), mode='lines+markers',  name='f(x)=sin(x)'))
fig.add_trace(go.Scatter(x=[-2*np.pi, 0, 2*np.pi, np.pi/2, -3*np.pi/2], y=[0, 0, 0, 1, 1], name='точки пересечения', mode='markers', marker=dict(color='Red', size=10)))
fig.add_trace(go.Scatter(x=x, y=g(x), mode='lines+markers',name='g(x)=1-cos(x)',
                         marker=dict(color='LightSkyBlue', size=3, line=dict(color='MediumPurple', width=3))))
fig.update_layout(legend_orientation="h",
                  legend=dict(x=.5, xanchor="center"),
                  hovermode="x",
                  title="График функции f(x)=sin(x) и g(x)=1-cos(x)",
                  xaxis_title="x",
                  yaxis_title="y",
                  margin=dict(l=0, r=0, t=40, b=0))
fig.update_traces(hoverinfo="all", hovertemplate="Аргумент: %{x}<br>Функция: %{y}")

fig.show()
```
